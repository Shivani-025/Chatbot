```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
tokenizer = AutoTokenizer.from_pretrained("valhalla/longformer-base-4096-finetuned-squadv1")
print("hiii")
```

```python
print("1hello")
def answer_question(question, text):
    print("2hello")
    encoding = tokenizer(question,text, return_tensors="pt")
    print("Encoding ",encoding)
    print("3hello")
    input_ids = encoding["input_ids"]
    print("Input id " , input_ids)
    #print("Input Ids")
    #print(input_ids)
    
    
    print("4hello")
# Get attention mask (local + global attention)
    attention_mask = encoding["attention_mask"]
    print("5hello")
    print(attention_mask)
# Get the predictions
    start_scores, end_scores = model(input_ids, attention_mask=attention_mask).values()
    print("6hello")
# Convert predictions into answer
    all_tokens = tokenizer.convert_ids_to_tokens(input_ids[0].tolist())
    print("7hello")
    '''for input_id in input_ids:
       token1 = all_tokens[input_id]
       print(f"Input ID: {input_id}, Token: {token1}")
    print("all_tokens ", all_tokens)'''
    
        
      
    answer_tokens = all_tokens[torch.argmax(start_scores):torch.argmax(end_scores)+1]
    print("8hello")
    #print(answer_tokens)
    answer = tokenizer.decode(tokenizer.convert_tokens_to_ids(answer_tokens))
    print("9hello")
    
    
    print("Ans:- "+answer)
    

print("10hello")
```
